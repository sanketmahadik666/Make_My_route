# Battery Data Access Demo

This notebook demonstrates how to use the `DataLoader` class to easily access standardized battery data.

```python
# Set up path to include src
import sys
import os
sys.path.append(os.path.abspath('..'))

from src.access.data_loader import DataLoader
```

```python
# Initialize DataLoader
loader = DataLoader(data_dir="../data/standardized/")

# Load the index
index = loader.load_index()
print("Available Files:")
display(index)
```

```python
# Filter for specific data
# Example: Get files for cell 'CS2_35'
files = loader.get_files({'cell_id': 'CS2_35'})
print(f"Found {len(files)} files for CS2_35")
```

```python
# Load a specific file
if files:
    df = loader.load_data(files[0])
    print(f"Loaded {len(df)} records from {files[0]}")
    display(df.head())
```
